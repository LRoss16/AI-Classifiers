```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
from sklearn import feature_extraction, model_selection, naive_bayes, metrics, svm
from IPython.display import Image
from keras.layers import SimpleRNN, Embedding, Dense, LSTM
from keras.models import Sequential
import seaborn as sns; sns.set()
%matplotlib inline  
```

```python
#Read in spam file
message=pd.read_csv('SMSSpamData.csv')
message.head()
#message.describe()
```

```python
#Graphs showing number of spam and ham messages
count_Class=pd.value_counts(message["class"], sort= True)
count_Class.plot(kind= 'bar', color= ["red", "black"])
plt.title('Ham vs Spam')
plt.show()

count_Class.plot(kind = 'pie',  autopct='%1.0f%%')
plt.title('Ham vs Spam Percentage')
plt.ylabel('')
plt.show()
```

```python
#Finding out the frequency of words in the messages
count1 = Counter(" ".join(message[message['class']=='ham']["sms_msg"]).split()).most_common(25)
hamGraph = pd.DataFrame.from_dict(count1)
hamGraph = hamGraph.rename(columns={0: "words in non-spam", 1 : "count"})
count2 = Counter(" ".join(message[message['class']=='spam']["sms_msg"]).split()).most_common(25)
spamGraph = pd.DataFrame.from_dict(count2)
spamGraph = spamGraph.rename(columns={0: "words in spam", 1 : "count_"})
```

```python
#Graphs detailing the most frequent words used in both spa and ham messages
hamGraph.plot.bar(legend = False, color = 'red')
y_pos = np.arange(len(hamGraph["words in non-spam"]))
plt.xticks(y_pos, hamGraph["words in non-spam"])
plt.title('More frequent words in ham messages')
plt.xlabel('words')
plt.ylabel('number')
plt.savefig( 'Non-spam', dpi = 300 )
plt.show()

spamGraph.plot.bar(legend = False, color = 'black')
y_pos = np.arange(len(spamGraph["words in spam"]))
plt.xticks(y_pos, spamGraph["words in spam"])
plt.title('More frequent words in spam messages')
plt.xlabel('words')
plt.ylabel('number')
plt.savefig( 'Spam', dpi = 300 )
plt.show()
```

```python
#Remove stop words
remove = feature_extraction.text.CountVectorizer(stop_words = 'english')
X = remove.fit_transform(message["sms_msg"])
np.shape(X)
```

```python
#Convert messages to binary form and split data into training and testing sets
message["class"]=message["class"].map({'spam':1,'ham':0})

X_train, X_test, y_train, y_test = model_selection.train_test_split(X, message['class'], test_size=0.2, random_state=42)
print([np.shape(X_train), np.shape(X_test)])
```

```python
#Using Multinominal naive bayes classifier

# Train different models changing the value alpha
#Evaluate the accuracy, recall and precision of model using test set
list_alpha = np.arange(1/100000, 20, 0.11)
score_train = np.zeros(len(list_alpha))
score_test = np.zeros(len(list_alpha))
recall_test = np.zeros(len(list_alpha))
precision_test= np.zeros(len(list_alpha))
count = 0
for alpha in list_alpha:
    bayes = naive_bayes.MultinomialNB(alpha=alpha)
    bayes.fit(X_train, y_train)
    score_train[count] = bayes.score(X_train, y_train)
    score_test[count]= bayes.score(X_test, y_test)
    recall_test[count] = metrics.recall_score(y_test, bayes.predict(X_test))
    precision_test[count] = metrics.precision_score(y_test, bayes.predict(X_test))
    count = count + 1 
```

```python
#show first 15 models and results
matrix = np.matrix(np.c_[list_alpha, score_train, score_test, recall_test, precision_test])
models = pd.DataFrame(data = matrix, columns = 
             ['alpha', 'Train Accuracy', 'Test Accuracy', 'Test Recall', 'Test Precision'])
models.head(n=15)
```

```python
#Select the model with best results
best_model = models['Test Precision'].idxmax()
models.iloc[best_model, :]
```

```python
#Check if other models have 100% test precision
models[models['Test Precision']==1].head(n=5)
```

```python
#Select model (out of ones with 100% precision) with best test accuracy
best_model = models[models['Test Precision']==1]['Test Accuracy'].idxmax()
bayes = naive_bayes.MultinomialNB(alpha=list_alpha[best_model])
bayes.fit(X_train, y_train)
models.iloc[best_model, :]
```

```python
#confusuin matrix used to test 
confusion_test = metrics.confusion_matrix(y_test, bayes.predict(X_test))
pd.DataFrame(data = confusion_test, columns = ['Bayes Predicted Ham', 'Bayes Predicted Spam'],
            index = ['Actual Ham', 'Actual Spam'])
```

```python
#Using SVM classifier

#train different models changing value C
list_C = np.arange(500, 2000, 100) 
score_train = np.zeros(len(list_C))
score_test = np.zeros(len(list_C))
recall_test = np.zeros(len(list_C))
precision_test= np.zeros(len(list_C))
count = 0
for C in list_C:
    svc = svm.SVC(C=C)
    svc.fit(X_train, y_train)
    score_train[count] = svc.score(X_train, y_train)
    score_test[count]= svc.score(X_test, y_test)
    recall_test[count] = metrics.recall_score(y_test, svc.predict(X_test))
    precision_test[count] = metrics.precision_score(y_test, svc.predict(X_test))
    count = count + 1 
```

```python
#View first 15 models and results

matrix = np.matrix(np.c_[list_C, score_train, score_test, recall_test, precision_test])
models = pd.DataFrame(data = matrix, columns = 
             ['C', 'Train Accuracy', 'Test Accuracy', 'Test Recall', 'Test Precision'])
models.head(n=15)
```

```python
#select model with best test precision

best_model = models['Test Precision'].idxmax()
models.iloc[best_model, :]
```

```python
#see if more than one model has 100% test precision
models[models['Test Precision']==1].head(n=5)
```

```python
#Select model (out of models with 100% test precision) which has best test accuracy
best_model = models[models['Test Precision']==1]['Test Accuracy'].idxmax()
svc = svm.SVC(C=list_C[best_model])
svc.fit(X_train, y_train)
models.iloc[best_model, :]
```

```python
#Use confusion matrix to test
confusion_test = metrics.confusion_matrix(y_test, svc.predict(X_test))
pd.DataFrame(data = confusion_test, columns = ['SVM Predicted Ham', 'SVM Predicted Spam'],
            index = ['Actual Ham', 'Actual Spam'])
```

```python
### Evaluating using Keras RNN classifier
```

```python
# Data from csv file is read and the messages are saved as list of string and corresponding labels and converted into numpy array 
messages = []
labels = []
for i, label in enumerate(message['class']):
    messages.append(message['sms_msg'][i])
    if label == 'ham':
        labels.append(0)
    else:
        labels.append(1)

messages = np.asarray(messages)
labels = np.asarray(labels)
```

```python
print("number of messages :" , len(messages))
print("number of labels: ", len(labels))
```

```python

from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences

#max number of words to consider
max_vocab = 10000
max_len = 500
tokenizer = Tokenizer(num_words=max_vocab)
#calucalte the frequency of each word
tokenizer.fit_on_texts(messages)
#converts array of strings of sequences to list of sequences of integers (1 is the most commonly used, 2 the 2nd most and so on)
sequences = tokenizer.texts_to_sequences(messages)

#convert list to 2d array
word_index = tokenizer.word_index
data = pad_sequences(sequences, maxlen=max_len)
shape = len(tokenizer.word_index)

print (shape)
```

```python
#Prepare test data: 80% for training, 20% for testing
train_samples = int(len(messages)*0.8)

messages_train = data[:train_samples]
labels_train = labels[:train_samples]

messages_test = data[train_samples:len(messages)-2]
labels_test = labels[train_samples:len(messages)-2]
#print (messages_test)
print (labels_test)
```

```python
#Build network

embedding_size=32
model = Sequential()
model.add(Embedding(input_dim=max_vocab,
 output_dim=embedding_size,
 input_length=max_len))
model.add(SimpleRNN(units=embedding_size))

model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer='rmsprop', loss='binary_crossentropy',
 metrics=['acc'])
model.summary()
```

```python
#Train and evaluate
model.fit(messages_train, labels_train, epochs=10, batch_size=60, validation_split=0.2)

acc = model.evaluate(messages_test, labels_test)
print("Test loss is {0:.2f} accuracy is {1:.2f} ".format(acc[0],acc[1]))
```

```python
#Test the classifier to make sure it is giving the correct prediction
#Test message should be classed as spam so output should be 1

def testNetwork(msg):
 msg = msg.lower().split(' ')
 test_classifier = np.array([word_index[word] for word in msg])
 test_classifier = np.pad(test_classifier, (500-len(test_classifier), 0),
 'constant', constant_values=(0))
 test_classifier = test_classifier.reshape(1, 500)
 return test_classifier
```

```python
spamMessage = 'you have been chosen to win a free phone click here to claim your prize'
test_classifier = testNetwork(spamMessage)
pred = model.predict_classes(test_classifier)
print(pred)
```
